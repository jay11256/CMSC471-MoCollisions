```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('data/Crash_Reporting.csv', low_memory=False)
df.head()
```

```python
# Convert Crash Date/Time to datetime objects
df['Crash Date/Time'] = pd.to_datetime(df['Crash Date/Time'])

# Extract time into a new column
df['Crash Time'] = df['Crash Date/Time'].dt.time
df[['Crash Date/Time', 'Crash Time']].head()
```

```python
# Select crashes from 12 AM (00:00) to 5 PM (17:00) 
# where severity is "No apparent injury" (case-insensitive)
time_mask = (df['Crash Date/Time'].dt.hour >= 0) & (df['Crash Date/Time'].dt.hour < 17)
severity_mask = df['Injury Severity'].str.contains('no apparent injury', case=False, na=False)

crashes_filtered = df[time_mask & severity_mask]

print(f"Number of 'No apparent injury' crashes from 12 AM to 5 PM: {len(crashes_filtered)}")
```
