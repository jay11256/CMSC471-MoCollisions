```python
"""
Preprocessing notebook for the MoCollisions dashboard.

This notebook is a thin wrapper around `preprocess.py` so the same
transformations are reproducible from either the command line or a notebook.

Output files (written to `data/`):
    crashes_clean.csv   - compact, numerically-encoded crash records
    summary.json        - dataset metadata + category dictionaries
    mc_zipcodes.geojson - Montgomery County ZIP polygons (with per-ZIP totals)
    zip_aggregate.json  - per-ZIP aggregate counts
"""
import pandas as pd
import numpy as np
```

```python
# Quick look at the raw data
df = pd.read_csv('data/Crash_Reporting.csv', low_memory=False)
print(df.shape)
df.head()
```

```python
# See how messy the raw category columns are
for col in ['Injury Severity', 'Weather', 'Light', 'Collision Type', 'Driver Distracted By']:
    print(f"\n=== {col} ===")
    print(df[col].value_counts().head(8))
```

```python
# Run the full preprocessing pipeline (idempotent - safe to re-run)
# Writes: data/crashes_clean.csv, data/summary.json,
#         data/mc_zipcodes.geojson, data/zip_aggregate.json
import subprocess, sys
subprocess.run([sys.executable, 'preprocess.py'], check=True)
```

```python
# Inspect the cleaned output
clean = pd.read_csv('data/crashes_clean.csv')
print(clean.shape)
print(clean.dtypes)
clean.head()
```
